# 1. Univariate Analysis

<b> A). To check which category has the highest number of sales. </b>

```python
fig, axs = plt.subplots(2,2, figsize = (15,11))
sns.countplot(x = d2['Category'],data = df, ax = axs[0,0], palette = 'Paired')
sns.countplot(x = d2['Segment'], ax = axs[0,1], data = df,palette = 'YlOrBr')
sns.countplot(x = d2['Ship Mode'], ax = axs[1,0], data = df,palette = 'magma_r')
sns.countplot(x = d2['Region'], ax = axs[1,1], data = df,palette = 'gist_earth_r')
```

<b> INSIGHTS :- </b>

<b> I. In the first plot, Office Supplies category was the most sold section.</b> <br>
<b> II.In the second plot, the Consumer segment goods were the most sold. </b> <br>
<b> III.In the third plot, the Standard class were the highest number of buyers. </b> <br>
<b> IV.In the fourth plot, the profit in West Region was the highest. </b>

<b> B). To check which state in the US has the highest number of sales.</b>

```python
plt.figure(figsize= (17,19))
sns.countplot(x = df['State'],data = df,palette =  'nipy_spectral_r')
plt.xticks(rotation=90)
plt.show()
```

<b> INSIGHTS :- </b> <br>
<b> I.) From the above plot it can be concluded that California has highest number of sales, followed by New York and then Texas. </b>

<b> C) To check how distributed are the numerical variables.</b>

```python
fig, axs = plt.subplots(2,2, figsize = (15,11))
sns.boxplot(d2['Sales'], ax = axs[0,0], palette = 'pink_r')
sns.boxplot(d2['Quantity'], ax = axs[0,1], palette = 'inferno_r')
sns.boxplot(d2['Discount'], ax = axs[1,0], palette = 'nipy_spectral')
sns.boxplot(d2['Profit'], ax = axs[1,1], palette = 'BuGn')
```

<b> INSIGHT :- From the above plots it can be concluded that we have negative profit too. </b>

# 2. Bivariate Analysis

<b> I. To figure out the goods sold (category wise) in respective regions of USA. </b>

```python
plt.figure(figsize= (13,10))
sns.countplot(x = d2['Region'],data = d2, hue = d2['Category'],palette = 'prism')
```

<b> From the above figure it can be concluded that West Region has highest number of sales in all categories. </b>

<b> II. To check whcich state has the highest profit. </b>

```python
v = d2.groupby(["State"])[['Profit']].sum()
m = v.sort_values(by = 'Profit',ascending = False)b
m.plot(kind = 'bar',figsize = (20,10), color = 'green')
plt.xticks(rotation = 90)
plt.show()
```

<b> INSIGHT :- From the above plot California witnessed the highest profit with a figure of 76,381.</b>

<b> III. To check which consumer segment was the most preffered among various economic groups. </b>

```python
plt.figure(figsize= (13,10))
sns.countplot(x = d2['Ship Mode'],hue = d2['Category'],data = d2, color = 'maroon')
```

<b> INSIGHT :- It was observed that 'Standard Class' was the highest buyer in all goods. </b>

<b> IV. To check which segment has the highest number of sales.</b>

```python
a = d2.groupby(["Segment"])[['Sales']].count()
b = a.sort_values(by = 'Sales',ascending = False)
b.plot(kind = 'bar',figsize = (11,10), color = 'red')
plt.xticks(rotation = 90)
plt.show()
```

<b> INSIGHT :- It was observed that Consumer commodities were sold the highest.</b>

<b> V. To check in which state sales was the highest </b>

```python
d2['State'].value_counts().plot(kind = 'bar',figsize = (22,17), color = 'Purple' )
plt.xlabel('States')
plt.ylabel('Sales')
plt.show()
```

<b> INSIGHT :- It was found that in California the sales were highest. </b>

<b> VI:- To check which region has the highest profit share. </b>

```python
p3 = d2.groupby('Region').sum()['Profit']
p3.plot.pie(autopct='%1.2f%%',figsize = (13,13))
```

<b> INSIGHT :- It was found that WEST region dominated the most in Profit share. </b>
